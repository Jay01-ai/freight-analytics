```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_all_freight.csv")
print(df.shape)
df.head()
```

```python
plt.figure(figsize=(10, 5))
df.groupby('CNSR')['TOTL FRGT(INCLUDE GST)'].sum().sort_values().plot(kind='barh', color='steelblue')
plt.title('Total Freight by Consignor')
plt.xlabel('Total Freight (₹)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
df.groupby('RR DATE')['TOTL FRGT(INCLUDE GST)'].sum().plot(kind='line', marker='o', color='green')
plt.title('Freight Value Over Time')
plt.xlabel('Date')
plt.ylabel('Total Freight (₹)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['STTN FROM'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title('Top 10 Source Stations')
plt.xlabel('Station')
plt.ylabel('Number of Shipments')
plt.tight_layout()
plt.show()
```

```python
gst_data = {
    'IGST': df['IGST'].sum(),
    'CGST': df['CGST'].sum(),
    'SGST': df['SGST'].sum(),
}
plt.figure(figsize=(6, 6))
plt.pie(gst_data.values(), labels=gst_data.keys(), autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('GST Breakdown')
plt.show()
```
