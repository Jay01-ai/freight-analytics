```python
import pandas as pd
import numpy as np

df = pd.read_csv("cleaned_all_freight.csv")
df['RR DATE'] = pd.to_datetime(df['RR DATE'])
print(df.shape)
```

```python
# Weight difference (charged vs actual)
df['WEIGHT_DIFF'] = df['CHRG WGHT'] - df['ACTL WGHT']
df['WEIGHT_DIFF_PCT'] = (df['WEIGHT_DIFF'] / df['ACTL WGHT']) * 100

# Freight per tonne
df['FREIGHT_PER_TONNE'] = df['TOTL FRGT(INCLUDE GST)'] / df['CHRG WGHT']

# GST percentage
df['GST_RATE_PCT'] = (df['TOTL GST'] / df['TOTL FRGT(EXCLUDE GST)']) * 100

# Binary flags
df['HAS_DEMURRAGE'] = (df['DMRG'] > 0).astype(int)
df['HAS_OTHER_CHARGES'] = (df['OTHR CHRG AMNT'] > 0).astype(int)

# Route column
df['ROUTE'] = df['STTN FROM'] + ' → ' + df['STTN TO']

# Date features
df['DAY_OF_WEEK'] = df['RR DATE'].dt.dayofweek
df['MONTH'] = df['RR DATE'].dt.month
df['WEEK'] = df['RR DATE'].dt.isocalendar().week.astype(int)

print("New columns added!")
print(df[['WEIGHT_DIFF','FREIGHT_PER_TONNE','GST_RATE_PCT','HAS_DEMURRAGE','ROUTE']].head())
```

```python
print("Shipments with demurrage:", df['HAS_DEMURRAGE'].sum())
print("Avg freight per tonne: ₹", round(df['FREIGHT_PER_TONNE'].mean(), 2))
print("Avg GST rate:", round(df['GST_RATE_PCT'].mean(), 2), "%")
print("Unique routes:", df['ROUTE'].nunique())
```

```python
df.to_csv("featured_freight.csv", index=False)
print("Saved as featured_freight.csv!")
print("Total columns now:", df.shape[1])
```
