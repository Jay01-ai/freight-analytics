```python
import pandas as pd

all_files = [
    "Outward Freight Details (4).csv",
    "Outward Freight Details (5).csv",
    "Outward Freight Details (6).csv",
    "Outward Freight Details (7).csv",
]

frames = []
for file in all_files:
    df = pd.read_csv(file)
    df = df[df['DVSN'].notna()].copy()
    df = df.drop_duplicates()
    df['RR DATE'] = pd.to_datetime(df['RR DATE'], dayfirst=True)
    df['INVC DATE'] = pd.to_datetime(df['INVC DATE'], dayfirst=True)
    num_cols = ['IGST','CGST','SGST','UGST','TOTL GST',
                'TDS IGST','TDS CGST','TDS SGST','TDS UGST','TOTL TDS',
                'DMRG','OTHR CHRG AMNT','REBATE AMNT']
    df[num_cols] = df[num_cols].fillna(0)
    df['_source'] = file   # tag which file each row came from
    frames.append(df)
    print(f"{file} → {len(df)} clean rows")

combined = pd.concat(frames, ignore_index=True)
print("\nAll files combined:", combined.shape)
combined.to_csv("cleaned_all_freight.csv", index=False)
print("Saved as cleaned_all_freight.csv!")
```
